### Sampling, Quantization and Coding of an analog signal

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 0.008, 0.00001)
x = 5 * np.sin(2 * np.pi * 500 * t + np.radians(90))

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(t, x)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('x(n) = 5sin(2π.500t + π/2)')

fs = 3750
n = np.arange(0, 0.008, 1/fs)
x_n = 5 * np.sin(2 * np.pi * 500 * n + np.pi / 2)
plt.subplot(2, 2, 2)
plt.plot(x_n)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n) = 5cos(2π.(500/3750)n)')

plt.subplot(2, 2, 3)
plt.stem(x_n)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n) = 5cos(2π.(500/3750)n)')

number_of_bits = 2
max = np.max(x_n)
min = np.min(x_n)
step_size = (max - min) / (2 ** number_of_bits - 1)
x_q = (np.round((x_n - min)/step_size)) * step_size + min

plt.subplot(2, 2, 4)
plt.stem(x_q, linefmt='--')
plt.plot(x_q)
plt.xlabel('n')
plt.ylabel('Amplitude (Quantized)')
plt.title('x(n) = 5cos(2π.(500/3750)n)')

print('The binary sequence of the given signal is:')
x_b = []
for i in x_q:
    x_b.append(bin(int((i - min) / step_size))[2:])
print(x_b)

plt.tight_layout()
plt.show()
```
