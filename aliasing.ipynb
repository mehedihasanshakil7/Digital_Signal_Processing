### Show that 50 Hz is an alias of the frequency 10 Hz, when sampling at 40 Hz. Assume the signals are
#### sin(2π.10t)
#### sin(2π.50t)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 1, 0.001)
x = np.sin(2 * np.pi * 10 * t)

plt.figure(figsize=(10, 4))

plt.subplot(2, 2, 1)
plt.plot(t, x)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('x(t) = sin(2π.10t)')

t = np.arange(0, 1, 0.001)
x = np.sin(2 * np.pi * 50 * t)

plt.subplot(2, 2, 2)
plt.plot(t, x)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('x(t) = sin(2π.50t)')

fs = 40
n = np.arange(0, 1, 1/fs)
x_n = np.sin(2 * np.pi * 10 * n)

plt.subplot(2, 2, 3)
plt.stem(n, x_n)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n) = sin(2π.(10/40)n)')

fs = 40
n = np.arange(0, 1, 1/fs)
x_n = np.sin(2 * np.pi * 50 * n)

plt.subplot(2, 2, 4)
plt.stem(n, x_n)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n) = sin(2π.(50/40)n)')

plt.tight_layout()
plt.show()
```
