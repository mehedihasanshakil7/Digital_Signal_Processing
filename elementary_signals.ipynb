```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sine Wave
t = np.arange(0, 1, 0.01)
y = 2 * np.sin(2 * np.pi * 2 * t)

plt.plot(t, y)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.show()
```

```python
# Cosine Wave
t = np.arange(0, 1, 0.01)
y = 2 * np.cos(2 * np.pi * 2 * t)

plt.plot(t, y)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Cosine Wave')
plt.show()
```

```python
# Unit Impulse Sequence
def unit_impulse(shift, n):
    delta = []
    for i in n:
        if(i == shift):
            delta.append(1)
        else:
            delta.append(0)
    return delta

# Main Function
n = np.arange(-5, 6, 1)
delta = unit_impulse(0, n)

plt.stem(n, delta)
plt.xlabel('n')
plt.ylabel('δ(n)')
plt.title('Unit Impulse Sequence')
plt.show()
```

```python
# Unit Step Sequence
def unit_step(shift, n):
    step = []
    for i in n:
        if(i < shift):
            step.append(0)
        else:
            step.append(1)
    return step

# Main Function
n = np.arange(-5, 6, 1)
step = unit_step(0, n)

plt.stem(n, step)
plt.xlabel('(n)')
plt.ylabel('u(n)')
plt.title('Unit Step Sequence')
plt.show()
```

```python
# Unit Ramp Sequence
def unit_ramp(shift, n):
    ramp = []
    j = 0
    for i in n:
        if(i < shift):
            ramp.append(0)
        else:
            ramp.append(j)
            j += 1
    return ramp

# Main function
n = np.arange(-5, 6, 1)
ramp = unit_ramp(0, n)

plt.stem(n, ramp)
plt.xlabel('n')
plt.ylabel('ur(n)')
plt.title('Unit Ramp Signal')
plt.show()
```

```python
# Exponential Signal, when 0<a<1
n = np.arange(0, 20, 1)
a = 0.8
exp = a ** n

plt.stem(n, exp)
plt.xlabel('n')
plt.ylabel('exp(n)')
plt.title('Exponential Signal (0<a<1)')
plt.show()
```
